```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
def matrix_get_submatrix(m, ii, jj):
    cols = len(m)     
    rows = len(m[0])  
    assert 0 <= ii < rows
    assert 0 <= jj < cols
        
    return [m[i][0:jj] + m[i][jj+1:] for i in range(rows) if i != ii]


def matrix_det(m):
    rows = len(m)     # number of rows of m
    assert rows > 0
    cols = len(m[0])  # number of cols of m
    if rows != cols:
        raise Exception("matrix must be square")
    if rows == 1:
        return m[0][0]
    elif rows == 2:
        return m[0][0] * m[1][1] - m[0][1] * m[1][0]
    elif rows == 3:
        return m[0][0]*m[1][1]*m[2][2] + m[0][1]*m[1][2]*m[2][0] + m[0][2]*m[1][0]*m[2][1] - m[2][0]*m[1][1]*m[0][2] - m[2][1]*m[1][2]*m[0][0] - m[2][2]*m[1][0]*m[0][1]    
     
    sum = 0
    for i in range(cols):
        if i % 2 == 0:
            sgn = 1
        else:
            sgn = -1
        sum += sgn * m[i][0] * matrix_det(matrix_get_submatrix(m, i, 0))
    return sum
```

```python
import pickle
import sys

from PyQt5.QtCore import QDir
from PyQt5.QtWidgets import QWidget, QLabel, QLineEdit, QTextEdit, QListView, QListWidget, QPushButton, QSpinBox, \
    QRadioButton, QButtonGroup, QAbstractItemView
from PyQt5.QtWidgets import QApplication, QDialog, QFileDialog
from PyQt5.QtWidgets import QGridLayout, QHBoxLayout, QVBoxLayout, QFormLayout, QMainWindow

from PyQt5.uic.properties import QtGui


class SubjectDialog(QDialog):

    def __init__(self, parent, data):
        super().__init__(parent)
        self.initUI(data)
        self.showDialog()


    def initUI(self, data):

        self.setGeometry(300, 300, 350, 300)
        self.setWindowTitle('Subject')
        self.show()

        self.name=QLabel('Name')
        self.line_name=QLineEdit(self)
        
        self.symptom=QLabel('Symptom')
        self.line_symptom=QLineEdit(self)
        
        self.year=QLabel('Year')
        self.line_year=QSpinBox(self)
        self.line_year.setRange(1900,2020)
        
        self.gender=QLabel('Gender')
        self.gender_male=QRadioButton('m',self)
        self.gender_female=QRadioButton('f',self)
        self.gender_unknown=QRadioButton('?',self)
        self.gendergroup = QButtonGroup()
        self.gendergroup.addButton(self.gender_male)
        self.gendergroup.addButton(self.gender_female)
        self.gendergroup.addButton(self.gender_unknown)
        
        hgbox=QHBoxLayout()
        hgbox.addWidget(self.gender_male)
        hgbox.addWidget(self.gender_female)
        hgbox.addWidget(self.gender_unknown)
        
        self.button_ok = QPushButton('Ok', self)
        self.button_cancel = QPushButton('Cancel', self)
        self.button_ok.clicked.connect(self.accept)
        self.button_cancel.clicked.connect(self.reject)

        horbox=QHBoxLayout()
        horbox.addWidget(self.button_ok)
        horbox.addWidget(self.button_cancel)
        
        subjectLayout = QFormLayout()
        subjectLayout.addRow(self.name, self.line_name)
        subjectLayout.addRow(self.year, self.line_year)
        subjectLayout.addRow(self.gender, hgbox)
        subjectLayout.addRow(self.symptom, self.line_symptom)
        subjectLayout.addRow(horbox)
        self.setLayout(subjectLayout)


    def getData(self):
        
        self.line_name.textChanged.connect(ListWindow.onEditClicked)
        self.line_year.valueChanged.connect(ListWindow.onEditClicked)
        self.line_symptom.textChanged.connect(ListWindow.onEditClicked)
        
        if self.gender_male.isChecked() == True:
            self.getgender = 'm'
        elif self.gender_female.isChecked() == True:
            self.getgender = 'f'
        elif self.gender_unknown.isChecked() == True:
            self.getgender = '?'
        
        return [str(self.line_name.text()), self.line_year.value(), self.getgender, str(self.line_symptom.text())]
        
    
    def showDialog(self):
        
        self.getname = self.line_name.text()
        self.getsymptom = self.line_symptom.text()
        self.getyear = self.line_year.value()

        
    def setData(self, data):
        self.line_name.setText(data[0])
        self.line_year.setValue(data[1])
        self.line_symptom.setText(data[3])
        
        if data[2] == 'm':
            self.gender_male.setChecked(True)
        elif data[2] == 'f':
            self.gender_female.setChecked(True)
        elif data[2] == '?':
            self.gender_unknown.setChecked(True)
        


class ListWindow(QMainWindow):

    def __init__(self):
        super().__init__()

        self.data = [['Lara', 1995, 'f', 'amputated leg'],
                     ['Faiza', 1998, 'f', 'Influenza'],
                     ['Marie', 1998, 'f', 'broken shoulder'],
                     ['Kurt', 1970, 'm', 'Corona'],
                     ['Alex', 1990, '?', 'cancer']]

        self.initUI()


    def initUI(self):
        
        self.button_add = QPushButton('Add', self)
        self.button_edit = QPushButton('Edit', self)
        self.button_delete = QPushButton('Delete', self)
        self.button_load = QPushButton('Load', self)
        self.button_save = QPushButton('Save', self)

        hbox = QHBoxLayout()
        hbox.addWidget(self.button_add)
        hbox.addWidget(self.button_edit)
        hbox.addWidget(self.button_delete)
        hbox.addWidget(self.button_load)
        hbox.addWidget(self.button_save)

        self.button_edit.clicked.connect(self.onEditClicked)
        self.button_delete.clicked.connect(self.onDeleteClicked)
        self.button_add.clicked.connect(self.onAddClicked)

        menuBar = self.menuBar()
        fileMenu = menuBar.addMenu('File')

        self.button_load.clicked.connect(self.get_text_file)
        self.button_save.clicked.connect(self.file_save)

        self.list = QListWidget()
        self.list.setAcceptDrops(True)
        self.list.setDragEnabled(True)
        self.list.setDragDropMode(QAbstractItemView.InternalMove)
        self.list.model().rowsMoved.connect(self.onRowsMoved)

        self.updateList()

        self.list.itemDoubleClicked.connect(self._handleDoubleClick)

        vbox = QVBoxLayout()
        vbox.addWidget(self.list)
        vbox.addLayout(hbox)

        centralWidget = QWidget(self)
        self.setCentralWidget(centralWidget)
        centralWidget.setLayout(vbox)

        self.setGeometry(300, 300, 350, 300)
        self.setWindowTitle('Subject List')
        self.show()


    def updateList(self):
        
        self.list.clear()
        for i in range(len(self.data)):
            self.list.addItem(self.data[i][0])
        return self.list

    
    def onEditClicked(self):

        current_row = self.list.currentRow()
        if current_row < 0:
            return

        self.setEnabled(False)

        dlg = SubjectDialog(self, self.data)

        dlg.setData(self.data[current_row])

        if dlg.exec_() == dlg.Accepted:
            self.data[current_row] = dlg.getData()
            self.updateList()

        self.setEnabled(True)

    def onDeleteClicked(self):
        
        current_row = self.list.currentRow()
        if current_row < 0:
            return
        self.list.takeItem(current_row)
        self.data.pop(current_row)
        return

    def _handleDoubleClick(self, item):
        
        self.onEditClicked()
        item.setSelected(False)

    def get_text_file(self):

        name = QFileDialog.getOpenFileName(self, 'Open File','c:\\',"Text files (*.txt)")
        print(name)
        file = open(name.__getitem__(0), 'rb')
        self.data = pickle.load(file)
        self.updateList()
        file.close()

    def file_save(self):
        
        name = QFileDialog.getSaveFileName(self, 'Save File','c:\\',"Text files (*.txt)")
        with open(name.__getitem__(0), 'wb') as f:
            pickle.dump(self.data, f)


    def onRowsMoved(self,parent, start, end, destination,row):
        
        #print ( "Moved rows!" ,  self.data[start], self.data[row])
        if start-row>=0:
            self.data.insert(row, self.data.pop(start))
            #print(self.data)

        else:
            self.data.insert(row-1, self.data.pop(start))
            #print(self.data)

    def onAddClicked(self):

        self.setEnabled(False)

        dlg = SubjectDialog(self, self.data)

        if dlg.exec_() == dlg.Accepted:
            self.data.append(dlg.getData())
            self.updateList()

        self.setEnabled(True)


if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = ListWindow()
    app.exec_()
```
